```python
import pandas as pd
```

```python
data=pd.read_csv('/workspaces/master_thesis/mapping/standard_concepts_with_synonyms_preprocessed.csv', on_bad_lines='skip')
```

```python
# delete columns where 'preprocessed' concept names is Nan
data=data[data.preprocessed.notna()]
```

```python
# check if for every preprocessed concept there is a vector from poincare model
# if not, print the concept_id of the concept

from gensim.models.poincare import PoincareModel
poincare_model=PoincareModel.load('/workspaces/master_thesis/poincare_100d_preprocessed')
```

```python
for row_id, row in data.iterrows():
    try:
        poincare_model.kv[row['preprocessed']]
    except:
        print(row['concept_id'])
```

```python
#delete concepts from data where concept_id did not have a vector in poincare model, concept_id = 45957596, 3667069, 45956745, 45769088, 37017139, 45956744
data=data[(data.concept_id!=45957596) & (data.concept_id!=3667069) & (data.concept_id!=45956745) & (data.concept_id!=45769088) & (data.concept_id!=37017139) & (data.concept_id!=45956744)]
```

```python
data_without_stemming=pd.read_csv('/workspaces/master_thesis/mapping/standard_concepts_with_synonyms_preprocessed_without_stemming.csv', on_bad_lines='skip')
```

```python
#data without stemming keep only concept_id which exist in data
data_without_stemming=data_without_stemming[data_without_stemming.concept_id.isin(data.concept_id)]
```

```python
data
```

```python
data_without_stemming=data_without_stemming[['preprocessed', 'preprocessed_synonyms']]
```

```python
#rename preprocessed to preprocessed_without_stemming and preprocessed_synonyms to preprocessed_synonyms_without_stemming
data_without_stemming=data_without_stemming.rename(columns={'preprocessed': 'preprocessed_without_stemming', 'preprocessed_synonyms': 'preprocessed_synonyms_without_stemming'})
```

```python
#merge data and data_without_stemming by index
data_merges=pd.merge(data, data_without_stemming, left_index=True, right_index=True)
```
