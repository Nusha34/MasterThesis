```python
import pandas as pd
from gensim.models import Word2Vec
import networkx as nx
```

```python
deepwalk_model = Word2Vec.load("/workspaces/master_thesis/deepwalk_snomed.model")
```

```python
result = deepwalk_model.wv.most_similar('4001098')
```

```python
result
```

```python
df_concept=pd.read_csv('/workspaces/master_thesis/CONCEPT.csv', on_bad_lines='skip', sep='\t')
```

```python
for i in range(len(result)):
    #concatenate dataframes with results 
    if i == 0:
        df_result_deepwalk = df_concept[df_concept['concept_id'] == int(result[i][0])]
    else:
        df_result_deepwalk = pd.concat([df_result_deepwalk, df_concept[df_concept['concept_id'] == int(result[i][0])]])
```

```python
df_result_deepwalk
```

```python
from gensim.models.poincare import PoincareModel
```

```python
poincare_model = PoincareModel.load(
        "/workspaces/master_thesis/poincare_100d_concept_id"
    )
```

```python
poincare_model.kv.most_similar(poincare_model.kv[4001098])
```

```python
result_poincare = poincare_model.kv.most_similar(poincare_model.kv[4001098], topn=6)
```

```python
result_poincare
```

```python
for i in range(len(result_poincare)):
    #concatenate dataframes with results 
    if i == 0:
        df_result_poincare = df_concept[df_concept['concept_id'] == result_poincare[i][0]]
    else:
        df_result_poincare = pd.concat([df_result_poincare, df_concept[df_concept['concept_id'] == result_poincare[i][0]]])
```
